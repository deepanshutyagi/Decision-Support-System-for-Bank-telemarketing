```python
#Author:Deepanshu Tyagi
#to get how many customers subscribed for loans
%matplotlib inline
import pandas as pd
h=pd.DataFrame.from_csv("bank-full.csv",index_col=None,encoding='utf8')
h.y.value_counts().plot(kind="pie",autopct='%1.1f%%',figsize=(6,6),label="Term deposit subscription")


```

```python
#Top job titles of people who subscribed for term deposits
x=h[h.y=="yes"]
x.job.value_counts().plot(kind="bar",figsize=(8,2.5))
```

```python
#bar graph for people who do management job and their education is known and have subscribed for term deposits 
# grouped by their eductaion, maritalstatus,housing loan,personal loan and past default prefrences.
x=h[(h.education!="unknown")&(h.job=="management")]
job_management=x[x.y=="yes"]
profile_management=job_management.groupby(["education","marital","housing","loan","default"])
profile_management.size().plot(kind="bar",figsize=(7,1.87),color="g")
```
